# Connect the datasets

```python
import feedparser as fp 
location = []
url = "https://www.paruvendu.fr/pa/rss/default/rssImmo"
reponse = fp.parse(url)
element = reponse.get("entries")
location.extend(element)
element
```

```python
dataframe_loc = pd.DataFrame.from_records(element)
dataframe_loc
```

```python
df_new_loc = pd.DataFrame()
df_new_loc["location"] = dataframe_loc["title"]
df_new_loc
```

```python
df_loc_final = pd.DataFrame()
df_loc_final['location'] = pd.DataFrame(list(df_new_loc.location.str.split("-")))[0]
df_loc_final
```

```python
df_final = pd.merge(data_frame_energy, df_loc_final, left_on="Nom commune", right_on="location")
df_final
```
