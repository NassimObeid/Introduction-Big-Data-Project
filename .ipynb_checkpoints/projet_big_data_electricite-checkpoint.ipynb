```python
# New data frame
data_frame_energy = pd.DataFrame()
```

```python
# Création de nouvelles colonnes
data_frame_energy['Année'] = data_frame['Année']
data_frame_energy['Nom commune'] = data_frame['Nom commune']
data_frame_energy['Nom département'] = data_frame['Nom département']
data_frame_energy['Code département'] = data_frame['Code département']
data_frame_energy['Conso totale Résidentiel (MWh)'] = data_frame['Conso totale Résidentiel (MWh)']
data_frame_energy['Consommation totale non résidents (MWh)'] = data_frame['Conso totale Professionnel (MWh)'] + data_frame['Conso totale Agriculture (MWh)'] + data_frame['Conso totale Industrie (MWh)'] + data_frame['Conso totale Tertiaire (MWh)'] + data_frame['Conso totale Secteur non affecté (MWh)']
```

# Functions: get data frame by city, departement filters

```python
def evolution_conso_residentiel_ville(df, ville):
    #print(type(df))
    #print(type(ville))
    #print(type(filtre))
    #print(filtre)
    
    edf = pd.DataFrame()
    
    if set(['Nom commune']).issubset(df.columns):
        edf = df.loc[df['Nom commune'] == ville];
        display(df.loc[df['Nom commune'] == ville])
        
        return edf;
    else:
        display("ERREUR")
```

```python
def evolution_conso_residentiel_departement(df, departement):
    #print(type(df))
    #print(type(ville))
    #print(type(filtre))
    #print(filtre)
    
    edf = pd.DataFrame()
    
    if set(['Code département']).issubset(df.columns):
        edf = df.loc[df['Code département'] == departement];
        display(df.loc[df['Code département'] == departement])
        
        return edf;
    else:
        display("ERREUR")
```

```python
# Pour sélectionner des lignes en fonction d'une valeur précise (d'une colonne ciblée)
# energie_data_frame.loc[energie_data_frame["Nom commune"] == 'Nanterre']
ville = 'Marseille'
filtre = "Nom commune"

edf = pd.DataFrame()
edf = evolution_conso_residentiel_departement(data_frame_energy, 1)

edfbis = pd.DataFrame();
edfbis['Conso totale Résidentiel (MWh)'] = edf['Conso totale Résidentiel (MWh)']
edfbis['Année'] = edf['Année']
edfbis.replace([np.inf, -np.inf], np.nan).dropna(axis=0)
# edf.dropna(axis=0, how='any')
# energie_data_frame.loc[energie_data_frame[filtre] == ville]
```
